```python
from openai import OpenAI
client = OpenAI(
    api_key = ""
)

weaviate_gorilla_prompt = """
An image featuring a massive, robot-like gorilla in a futuristic, industrial setting. The gorilla is highly detailed, with intricate mechanical parts and wiring, giving it a cybernetic appearance. Its body structure is made up of segmented, metallic components and reinforced joints, suggesting strength and durability. The gorilla's chest prominently displays a glowing, green "W," which stands out against its monochromatic, metallic body.

The background consists of dark, tiled walls with a grid pattern, illuminated by linear lights that add a sleek, modern feel to the setting. The gorilla is positioned in a low, powerful stance, conveying a sense of readiness or dominance, and the lighting accentuates the texture and details on its body, enhancing its imposing presence. The overall aesthetic combines elements of industrial design, futurism, and high-tech robotics.
"""

response = client.images.generate(
  model="dall-e-3",
  prompt=weaviate_gorilla_prompt,
  size="1024x1024",
  quality="standard",
  n=1,
  response_format="b64_json"
)

image_url = response.data[0]
```

```python
import base64
from PIL import Image
import io

# Decode base64 string to bytes
image_bytes = base64.b64decode(image_url.b64_json)

# Create PIL Image from bytes
image = Image.open(io.BytesIO(image_bytes))

# Display the image
display(image)
```
