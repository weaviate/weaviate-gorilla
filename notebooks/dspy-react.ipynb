```python
import dspy
lm = dspy.LM('openai/gpt-4o-mini', api_key='')
dspy.configure(lm=lm)

lm("say hello")
```

```python
from typing import Optional

def evaluate_math(expression: str, precision: Optional[int] = None):
    result = dspy.PythonInterpreter({}).execute(expression)
    if precision is not None:
        return round(result, precision)
    return result

react = dspy.ReAct("question -> answer: float", tools=[evaluate_math])

# Test with optional arguments
pred = react(question="What is 123.456789 * 2? Round to 2 decimal places.")
print(pred.answer)

# Test without optional arguments
pred2 = react(question="What is 123.456789 * 2?")
print(pred2.answer)
```
